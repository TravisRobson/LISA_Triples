```python
true = np.genfromtxt("true_triple.dat")
f    = true[:,0]
xRe  = true[:,1]
xIm  = true[:,2]
xPow = np.sqrt(xRe**2 + xIm**2)

plt.figure(figsize=(8, 6))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.xlabel(r'$f$ [Hz]', fontsize=20)
plt.ylabel(r'$\tilde{h}$', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)

plt.semilogy(f, xPow, label='Triple')

mx   = np.genfromtxt("max_post_src.dat")
f    = mx[:,0]
xRe  = mx[:,1]
xIm  = mx[:,2]
xPow = np.sqrt(xRe**2 + xIm**2)

plt.semilogy(f, xPow, label='Max. Post.')

# plt.savefig('small.pdf')
plt.legend(loc='lower left', prop={'size': 24} )
plt.show()
```

```python
true = np.genfromtxt("test_signal.dat")
f = true[:,0]
xRe = true[:,1]
xIm = true[:,2]
xPow = np.sqrt(xRe*xRe + xIm*xIm)

mx = np.genfromtxt("max_signal.dat")
f_max = mx[:,0]
xRe_max = mx[:,1]

xIm_max = mx[:,2]
xPow_max = np.sqrt(xRe_max**2 + xIm_max**2)
```
